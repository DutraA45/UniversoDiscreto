```python
%matplotlib inline
import preprocessing as pre
import regressionlinear as rl
import regressionpoly as plr

import numpy as np
import pandas as pd
import time
from functools import wraps
```

```python
#https://towardsdatascience.com/machine-learning-polynomial-regression-with-python-5328e4e8a386
#https://medium.com/coinmonks/polynomial-regression-11bec9262d64
```

```python
def runPolynomialRegressionExample(filename):
    start_time = time.time()
    X, y, csv = pre.loadDataset(filename)
    elapsed_time = time.time() - start_time
    print("Load Dataset: %.2f" % elapsed_time, "segundos.")
    
#    start_time = time.time()
#    X = pre.fillMissingData(X, 0, 1)
#    elapsed_time = time.time() - start_time
#    print("Fill Missing Data: %.2f" % elapsed_time, "segundos.")
    
    start_time = time.time()
    linearRegressor = rl.computeLinearRegressionModel(X, y)
    elapsed_time = time.time() - start_time
    print("Compute Linear Regression: %.2f" % elapsed_time, "segundos.")
    
    start_time = time.time()
    XPoly, polyLinearRegressor = plr.computePolynomialLinearRegressionModel(X, y, 4)
    elapsed_time = time.time() - start_time
    print("Compute Polynomial Linear Regression: %.2f" % elapsed_time, "segundos.")
    
    rl.showPlot(X, y, linearRegressor)
    plr.showPlot(X, y, X, polyLinearRegressor.predict(XPoly))
```

```python
if __name__ == "__main__":
    runPolynomialRegressionExample("salary.csv")
```
