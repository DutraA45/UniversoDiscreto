```python
%matplotlib inline
import preprocessing as pre
import svr as svr

import numpy as np
import pandas as pd
import time
from functools import wraps
```

```python
#https://www.kdnuggets.com/2017/03/building-regression-models-support-vector-regression.html
#https://medium.com/coinmonks/support-vector-regression-or-svr-8eb3acf6d0ff
#http://prorum.com/?qa=2051/como-funcionam-os-modelos-de-support-vector-regression
#https://www.thedynamatrix.com/machine-learning/support-vector-regression
```

```python
def runSupportVectorRegressionExample(filename):
    start_time = time.time()
    X, y, csv = pre.loadDataset(filename)
    elapsed_time = time.time() - start_time
    print("Load Dataset: %.2f" % elapsed_time, "segundos.")
    
    start_time = time.time()
    X, scaleX = pre.computeScaling(X)
    y, scaleY = pre.computeScaling(np.reshape(y, (-1, 1)))
    elapsed_time = time.time() - start_time
    print("Compute Scaling: %.2f" % elapsed_time, "segundos.")
    
    start_time = time.time()
    svrModel = svr.computeSupportVectorRegressionModel(X, y, "linear", 2)
    elapsed_time = time.time() - start_time
    print("Compute SVR with kernel Linear: %.2f" % elapsed_time, "segundos.")
    
    svr.showPlot(scaleX.inverse_transform(X), scaleY.inverse_transform(y), scaleX.inverse_transform(X), scaleY.inverse_transform(svrModel.predict(X)))
    
    start_time = time.time()
    svrModel = svr.computeSupportVectorRegressionModel(X, y, "poly", 3)
    elapsed_time = time.time() - start_time
    print("Compute SVR with kernel Poly: %.2f" % elapsed_time, "segundos.")
    
    svr.showPlot(scaleX.inverse_transform(X), scaleY.inverse_transform(y), scaleX.inverse_transform(X), scaleY.inverse_transform(svrModel.predict(X)))
    
    start_time = time.time()
    svrModel = svr.computeSupportVectorRegressionModel(X, y, "rbf", 2)
    elapsed_time = time.time() - start_time
    print("Compute SVR with kernel RBF: %.2f" % elapsed_time, "segundos.")
    
    svr.showPlot(scaleX.inverse_transform(X), scaleY.inverse_transform(y), scaleX.inverse_transform(X), scaleY.inverse_transform(svrModel.predict(X)))
```

```python
runSupportVectorRegressionExample("salary.csv")
```
