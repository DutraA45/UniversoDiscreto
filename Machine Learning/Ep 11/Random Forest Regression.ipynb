```python
%matplotlib inline
import preprocessing as pre
import regressionrandomforest as rf

import numpy as np
import pandas as pd
import time
from functools import wraps
```

```python
#https://github.com/lucaslattari/UniversoDiscreto/tree/master/Machine%20Learning
#https://stackabuse.com/random-forest-algorithm-with-python-and-scikit-learn/
```

```python
def runRandomForestRegressionExample(filename):
    start_time = time.time()
    X, y, csv = pre.loadDataset(filename)
    elapsed_time = time.time() - start_time
    print("Load Dataset: %.2f" % elapsed_time, "segundos.")
    
    start_time = time.time()
    rfModel = rf.computeRandomForestRegressionModel(X, y, 100)
    elapsed_time = time.time() - start_time
    print("Compute Random Forest Regression: %.2f" % elapsed_time, "segundos.")
    
    rf.showPlot(X, y, X, rfModel.predict(X))
    
    XGrid = np.arange(min(X), max(X), 0.01)
    XGrid = XGrid.reshape((len(XGrid), 1))
    rf.showPlot(X, y, XGrid, rfModel.predict(XGrid))
```

```python
runRandomForestRegressionExample("salary2.csv")
```
